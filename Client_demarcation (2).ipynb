```python
import pandas as ed
Data = ed.read_csv("/content/CS Review (1).csv")
```

```python
import numpy as np
```

```python
Data
```

```python
Data = Data.fillna(0)
```

```python
# First quartile (Q1)
ARR_Q1 = np.percentile(Data['ARR'], 25,interpolation = 'midpoint')
```

```python
ARR_Q2 = np.percentile(Data['ARR'], 50, interpolation= 'midpoint')
```

```python
ARR_Q3 = np.percentile(Data['ARR'], 75, interpolation= 'midpoint')
```

```python
CompanySize_Q1 = np.percentile(Data['Company Size'], 25,interpolation = 'midpoint')
```

```python
CompanySize_Q2 = np.percentile(Data['Company Size'], 50,interpolation = 'midpoint')
```

```python
CompanySize_Q3 = np.percentile(Data['Company Size'], 75,interpolation = 'midpoint')
```

```python
# This code transforms data from the 'Data' DataFrame and assigns weightage values to different variables.
# It populates 'CompanySize_List', 'ARR_List', 'Region_List' and 'Feature_List' based on specific conditions.
ARR_List = []
Feature_List = []
CompanySize_List = []
for i in range(len(Data)) :
  if Data.loc[i]['Company Size'] == 0 :
    CompanySize_List.append(0)
  elif Data.loc[i]['Company Size'] < CompanySize_Q1 or  Data.loc[i]['Company Size'] == CompanySize_Q1 :
    CompanySize_List.append('.50')
  elif Data.loc[i]['Company Size'] > CompanySize_Q1 or  Data.loc[i]['Company Size'] == CompanySize_Q1 and  Data.loc[i]['Company Size'] < CompanySize_Q2 :
    CompanySize_List.append('1.00')
  elif Data.loc[i]['Company Size'] > CompanySize_Q2 or  Data.loc[i]['Company Size'] == CompanySize_Q2 and  Data.loc[i]['Company Size'] < CompanySize_Q3 :
    CompanySize_List.append('1.50')
  elif Data.loc[i]['Company Size'] > CompanySize_Q3 :
    CompanySize_List.append('2.00')
  if Data.loc[i]['ARR'] == 0 :
     ARR_List.append(0)
  elif Data.loc[i]['ARR'] < ARR_Q1 or  Data.loc[i]['ARR'] == ARR_Q1:
    ARR_List.append('1')
  elif Data.loc[i]['ARR'] > ARR_Q1 or  Data.loc[i]['ARR'] == ARR_Q1 and Data.loc[i]['ARR'] < ARR_Q2 :
    ARR_List.append('2')
  elif Data.loc[i]['ARR'] > ARR_Q2 or  Data.loc[i]['ARR'] == ARR_Q2 and Data.loc[i]['ARR'] < ARR_Q3 :
    ARR_List.append('3')
  elif Data.loc[i]['ARR'] > ARR_Q3 :
    ARR_List.append('4')
  if Data.loc[i]['Feature Count'] == 0 :
    Feature_List.append(0)
  elif Data.loc[i]['Feature Count'] == 0 or Data.loc[i]['Feature Count']  == 1 :
    Feature_List.append('.75')
  elif Data.loc[i]['Feature Count'] == 2 :
    Feature_List.append('1.5')
  elif Data.loc[i]['Feature Count']  == 3 :
    Feature_List.append('2.25')
  elif Data.loc[i]['Feature Count'] > 3 :
    Feature_List.append('3.00')


```

```python
Region_List = []
for i in range(len(Data)) :
  if i == 0 :
    Region_List.append(0)
  if Data.loc[i]['Region'] == 'India' :
    Region_List.append(1)
  elif Data.loc[i]['Region'] == 'Outside India' :
    Region_List.append(2)


```

```python
Data['ARR_class'] = ARR_List
Data['Feature_class'] = Feature_List
Data['CompanySize_class'] = CompanySize_List
Data['Region_Class'] = Region_List

```

```python
aggregated_values = []
sum_list = []
for i in range(len(Data)):
    arr = int(ARR_List [i])
    comp = float(CompanySize_List[i])
    feature = float(Feature_List [i])
    Region = float(Region_List[i])
    s = float(arr+comp+feature+Region)
    sum_list.append(s)
    if s == 0 :
      aggregated_values.append('null')
    if s < 3 and s == 3:
      aggregated_values.append('Bronze')
    elif s > 3 and s <= 5 :
      aggregated_values.append('Silver')
    elif s > 5 and s <= 6 :
      aggregated_values.append('Gold')
    elif s > 6 :
      aggregated_values.append('Platinum')
# The code above calculates the sum of values from different lists and assigns categories based on the sum.
# It iterates over the data and performs the necessary calculations and comparisons.
# The assigned categories are stored in the 'aggregated_values' list.
# The sum of values is stored in the 'sum_list' for reference or further analysis.

```

```python
Data['Sum_list'] = sum_list
```

```python
Data['Category'] = aggregated_values
```
